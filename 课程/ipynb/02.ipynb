# 1. 第几天？



```python

date = input("请输入年/月/日:")
```

```python
day = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
```

```python
def my_day(date):
    if int(date[:4]) % 4 == 0:
        if int(date[:4]) % 100 == 0:
            if int(date[:4]) % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False


```

```python
my_day(date)
```

```python
def my_day2(date):
    return sum(day[:1]) + sum(day[2:5]) + (lambda x: 29 if my_day(x) else 28)(date) + int(date[8:10])
```

```python
print("这是{}年的第{}天".format(date[:4], my_day2(date)))
```

# 2. 输出菱形图案



```python
n = int(input("请输入一个奇数"))
for i in range(-int(n / 2), int(n / 2) + 1):
    print(" " * abs(i), "*" * abs(n - abs(i) * 2))
```

# 3.数字陷阱




```python
Num = int(input("请输入一个30000下的正数:"))
```

```python
def Fanc(Num):
    W = int(Num / 10000)
    Q = int((Num - (W * 10000)) / 1000)
    B = int((Num - (W * 10000) - (Q * 1000)) / 100)
    S = int((Num - (W * 10000) - (Q * 1000) - (B * 100)) / 10)
    G = int(Num - (W * 10000) - (Q * 1000) - (B * 100) - (S * 10))
    return (W + Q + B + S + G) * 3 + 1
```

```python
i = Fanc(Num)
while Fanc(i) != i:
    i = Fanc(i)
    print(i)
else:
    print(i)
```

# 4.计算圆周率



```python
def molecule(n):
    if n == 0:
        return 1
    else:
        return n * molecule(n - 1)


def denominator(n):
    if n == 1:
        return 1
    else:
        return n * denominator(n - 2)
```

```python
def print_pi(Num):
    i = 1
    Sum = 1
    while True:
        if (molecule(i) / denominator(2 * i + 1)) > Num:
            Sum += (molecule(i) / denominator(2 * i + 1))
        else:
            return Sum * 2
        i += 1
```

```python
Num = float(input("请输入阈值:"))
```

```python
print("{:.6}".format(print_pi(Num)))
```
